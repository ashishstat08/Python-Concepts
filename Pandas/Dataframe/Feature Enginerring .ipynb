### We need to create some new features for that we need to take a copy of our data 


```python
uber2=uber.copy()
```

```python
uber2.rename(columns={'START_DATE*':'start_date',
              'END_DATE*': 'end_date',
              'CATEGORY*': 'category',
            'START*':'start',
            'STOP*':'stop',
            'MILES*':'miles',
            'PURPOSE*':'purpose'},inplace=True)
```

```python
uber2.head()
len(uber2)
```

```python
uber_n=uber2[:-1]
print(len(uber_n))
uber_n.info()
```

```python
# to convert the date time 
import datetime 
def convert(col_name):
    y=[]
    for x in uber_n[col_name]:
        y.append(datetime.datetime.strptime(x,"%m/%d/%Y %H:%M"))
    uber_n[col_name]=y
    
col_name=uber_n[['start_date','end_date']]
for x in col_name:
    convert(x)
```

```python
uber_n.info()
```

# create the new coloumn called month

```python
count=0
mont=[]
while count<len(uber_n):
    mont.append(uber_n['start_date'][count].month)
    count=count+1
uber_n['month']=mont
```

```python
uber_n['month2']=uber_n['start_date'].map(lambda x: x.month)
```
