# 1. Group by operation 


```python
import numpy as np 
import pandas as pd 
import matplotlib as plt 
import os

data ={'Company':['GOOG', 'GOOG', 'MICR', 'MICR', 'FB', 'FB'],
       'Person':['Sam', 'Charli', 'Chaplin', 'Mahendra', 'Singh', 'Dhoni'],
       'Sales':[200,399,400,200,500,444]}

data_new=pd.DataFrame(data)
data_new
```

```python
bycompy=data_new.groupby(['Company'])
bycompy.mean()
```

```python
# this will return column inside a column just like heir indexing in rows
by_new=bycompy.describe()
by_new
```

```python
by_new.loc[['GOOG','MICR']]['Sales'].iloc[:,1:3]
by_new['Sales'].loc[['GOOG','MICR'],['mean','std']]
```

```python
by_new.loc[['GOOG','MICR']].iloc[:,1:3]
```

```python
# you can also transpose the data 
tra_new=by_new.transpose()
tra_new.loc['Sales'].loc[['mean'],['FB','GOOG']]
```

# Merging , joinning and concating the dataset 


```python
df1=pd.DataFrame({'A':['A0','A1','A2','A3','A4'],
                 'B':['B0','B1','B2','B3','B4'],
                 'C':['C0','C1','C2','C3','C4']},index=[0,1,2,3,4])
```

```python
df2=pd.DataFrame({'A':['A5','A6','A7','A8','A9'],
                 'B':['B5','B6','B7','B8','B9'],
                 'C':['C5','C6','C7','C8','C9']},index=[0,1,2,3,4])
```

```python
df3=pd.DataFrame({'A':['A10','A11','A12','A13','A14'],
                 'B':['B10','B11','B12','B13','B14'],
                 'C':['C10','C11','C12','C13','C14']},index=[5,6,7,8,9])
```

# Concatenation

```python
# row wise
df_ct1=pd.concat([df1,df2,df3],axis=0)
df_ct1
```

```python
# column wise
df_ct1=pd.concat([df1,df2,df3],axis=1)
df_ct1
```

```python
df_ct1.fillna(value=0,inplace=True)
df_ct1
```

# Merge - this is similar to join but it will take column to join

```python
left=pd.DataFrame({'Key':['K0','K5','K1','K2','K3'],
                 'B':['B0','B1','B2','B3','B4'],
                 'C':['C0','C1','C2','C3','C4']},index=[0,1,2,3,4])
left.head()
```

```python
right=pd.DataFrame({'Key':['K0','K1','K2','K3','K4'],
                 'B':['B5','B6','B7','B8','B9'],
                 'C':['C5','C6','C7','C8','C9']},index=[0,1,2,3,4])
right.head()
```

```python
df_mer=pd.merge(left,right,how='inner',on='Key')
df_mer
```

## joins - simipler than merge but it takes indexs to join the two tables 

```python
left1=pd.DataFrame({'A':['A0','A1','A2','A3','A4'],
                 'B':['B0','B1','B2','B3','B4'],
                 'C':['C0','C1','C2','C3','C4']},index=[0,1,2,3,4])
```

```python
right1=pd.DataFrame({'AA':['A5','A6','A7','A8','A9'],
                 'BB':['B5','B6','B7','B8','B9'],
                 'CC':['C5','C6','C7','C8','C9']},index=[0,1,5,3,4])
```

```python
left1.join(right1,how='outer')
```
