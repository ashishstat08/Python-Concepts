# Joinning the two data frames in python

```python
users
```

```python
trans
```

```python
# joinning users and transcation 
ljoin=pd.merge(trans,users,left_on='UserID',right_on='UserID',how='left')
ljoin
```

```python
# transcation id not in users 
tr=np.where(ljoin['UserID'].isin(ljoin['TransactionID']))
tr
```

```python
#Join users to transactions, keeping only rows from transactions and users that match via UserID (inner join)
inner=pd.merge(trans,users,on='UserID',how='inner')
inner
```

```python
#Join users to transactions, displaying all matching rows AND all non-matching rows (full outer join)
outer=pd.merge(trans,users,on='UserID',how='outer')
outer
```
