# Pie Chart

> - autopct = it is used to show the percentage of the number covered


```python
iris=sns.load_dataset('iris')
iris.head()

df = iris.drop(['species'],axis =1)
data = df.iloc[0]
data
```

```python
#simple plot 
data.plot.pie(figsize = (7,7))
```

```python
# to make the subplots of the data 
data2 = df .head(3).T
```

```python
data2.plot.pie(subplots = True , figsize =(20,20))
```

```python
# to get the percentage of the plot 
data2.plot.pie(subplots = True , figsize =(20,20), autopct = '%.2f')
```
