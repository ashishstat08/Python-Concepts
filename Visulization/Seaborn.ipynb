# Seaborn 
> - sns.set = to set the theme of the graph

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib  inline 
```

```python
sns.set(style = 'darkgrid')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## 1. relplot
> - hue =  Grouping variable that will produce elements with different colors
> - size = size of the particular group i.e depend upon the data 
> - style 
> - ci = sd gives the standard deviation of the data distribution in form of highlighted
> - marker will give you the dots in the line i.e true or false
> - col = if you have different categorical variable than it will give the graph using each single category
> - col_wrap = graphs  at the single place i.e max no of graphs 
> - height - of the graph

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips)
```

```python
# re=tips[tips['smoker'] == 'No']
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips, hue = 'smoker', col = 'smoker', col_wrap=3, height=3 )
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips, hue = 'smoker',style ='time', size = 'size',)
```

# 1. catplot

```python
data = sns.load_dataset('tips')
data.head()
```

# joint plot 


> - c is the color for the marker 
> - s is the size
> - linewidth is the line of the distribution


```python
x = tips['total_bill']
y = tips ['tip']
```

```python
sns.jointplot(x=x , y = y , color = 'g')
```

```python
g = sns.jointplot(x =x , y= y , kind = 'kde', color = 'g')
g.plot_joint(plt.scatter, c = 'g', s =30, linewidth = 7 , marker = '+')
```
