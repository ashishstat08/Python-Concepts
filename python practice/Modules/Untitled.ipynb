# random modules

```python
import random as r

#1
r.choice([1,2,3,4,5])

#2
r.randrange(12,100,2)

#3 it will write the rndom number map with the seeded value 
r.seed(2)
r.random()

#shuffle the number 

li=[1,3,6,7,4,4]
for i in range(0,len(li)):
    print(li[i],end=" ")
print("\r")

random.shuffle(li)   
for i in range(0,len(li)):
    print(li[i],end=" ")
print("\r")

print ("The random floating point number between 5 and 10 is : ",end="") 
print (random.uniform(5,9))

```
